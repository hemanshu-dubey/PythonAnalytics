```
from pyspark.sql import Row
def f(s):
  if s.slab.strip() == '<=50K':
    return 0
  else:
    return 1
  
new_slab = income.select('slab').rdd.map(f).map(lambda x: Row(slab_cat=x)).toDF()
```
