```
x = sc.parallelize([("a", 1), ("b", 1), ("a", 1),("a", 1), ("b", 1), ("a", 1),("a", 1), ("b", 1), ("a", 1)],2)
def add(a, b): 
  print 'a ',a,type(a)
  print 'b ',b,type(b)
  return a + str(b)

def add1(a, b): 
  print 'add1 a ',a,type(a)
  print 'add1 b ',b,type(b)
  return a + str(b)

sorted(x.combineByKey(str, add, add1).collect())
```

```
rdd = sc.parallelize([1, 2, 3, 4], 2)
def f(iterator): yield sum(iterator)
rdd.mapPartitions(f).collect()
```

```
rdd = sc.parallelize([1, 2, 3, 4], 2)
def f(splitIndex, iterator): yield (splitIndex,sum(iterator))
rdd.mapPartitionsWithIndex(f).collect()
```
