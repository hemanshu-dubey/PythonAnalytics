Creating Dataframe for vertices & edges.

```
v = sqlContext.createDataFrame([
  ("a", "Alice", 34),
  ("b", "Bob", 36),
  ("c", "Charlie", 30),
  ("d", "David", 29),
  ("e", "Esther", 32),
  ("f", "Fanny", 36),
  ("g", "Gabby", 60)
], ["id", "name", "age"])
# Edge DataFrame
e = sqlContext.createDataFrame([
  ("a", "b", "friend"),
  ("b", "c", "follow"),
  ("c", "b", "follow"),
  ("f", "c", "follow"),
  ("e", "f", "follow"),
  ("e", "d", "friend"),
  ("d", "a", "friend"),
  ("a", "e", "friend")
], ["src", "dst", "relationship"])
```

```
from graphframes import *

g = GraphFrame(v, e)
```

```
g.degrees.collect()
```

```
g.inDegrees.collect()
```

```
g.vertices.collect()
```

```
g.edges.collect()
```

```
g.vertices.groupby().min("age").show()
```

```
from pyspark.sql.functions import *
g.vertices.select(min('age')).show()
```

```
numFollows = g.edges.filter("relationship = 'follow'").count()
```

```
numFollows
```

```
motifs = g.find("(a)-[e]->(b); (b)-[e2]->(a)")
motifs.show()
```

```
g.vertices.show()
```

```
g.bfs(fromExpr = "id = 'a'", toExpr = "id = 'b'")
```

```
g.bfs("name = 'Esther'","name = 'Bob'")
```
