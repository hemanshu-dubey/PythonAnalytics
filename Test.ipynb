```
from pyspark import Row
l = [ Row(word='one'), Row(word='two'), Row(word='one')]
```

```
df = spark.createDataFrame(l)
df.show()
```

```
def f(x):
  if x.word == 'one':
    d = 1
  else:
    d = 0
  #Add name=d in each Row, Row should retain earlier content
  #x.asDict() - convert Row into dictionary
  #new Row object is created with new column contents
  return Row(name=d,**x.asDict())

df = df.rdd.map(f).toDF()
```

```
df.show()
```

```
def f(x):
  if x.word == 'one':
    d = 1
  else:
    d = 0
  return Row(inf=d,**x.asDict())
df = df.rdd.map(f).toDF()
```

```
df.show()
```

```
df.collect()
```
